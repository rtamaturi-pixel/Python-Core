## for Loop:
- A for loop is used to **Iterate over a sequence** (like List, Tuple, Set, Dictionary, or Range.)

```python
s = "python"
for i in s:         #takes the value of sequence one by one
    print(i) 
```

```python
for i in s:
    print(i,end=' ')     #consider end with space.
```

```python
pos = 0
neg = 0
zero = 0

for num in numbers:
    if num > 0:
       pos += 1
    elif num < 0:
        neg +=1
    else:
        zero += 1
print("count of positive number:", pos)
print("count of negative number:", neg)
print("count of zero number:", zero)
```

## While Loop:

- A while Loop is used to repeat a block of code, as **long as the condition is True**.
- It is mainly used when **we don't know how many times we need to run the loop**.
- It is used when we want to keep repeating **until a specific condition is met.**


```python
i = 1
while i <= 20:
    print(i,end=' ')
    i += 1
```

```python
i = 1
while i <= 20:
   if i % 2 == 0:
       print(i,end=' ')
   i = i + 1
```

```python
sum_n = 0
i = 1
while i <= 10:
    sum_n += i
    i += 1
print(sum_n)
    
```

```python
num = int(input("Enter the number:"))
count = 0
while num > 0:
    count += 1
    num = num // 10       # floor divison used to count the interger part.
print("Number of digitis:", count)
```

```python
num = int(input("Enter the number:"))
rev = 0

while num > 0:
    digit = num % 10
    rev = rev *10 + digit
    num = num // 10
print("the reverse of the number is :", rev)
```

```python
num = int(input("Enter the number:"))
S= 0
Temp = num
while num > 0:
    digit = num % 10
    S = S + (digit**3)
    num = num // 10
if S == Temp:
    print("The given number is Armstrong")
else:
    print("The given number is not Armstrong")
```
