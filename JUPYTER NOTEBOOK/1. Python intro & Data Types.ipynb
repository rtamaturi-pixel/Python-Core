### Identifiers

1. Name of a variable / class / object / method is called an Identifier 
2. Identifiers are used to write meaningful instructions to the computer along with keyword


### Identifier Name Rules
- Identifier should **only start with a character or an underscore ( _ )**
- Identifier should **only contain alphanumerics and an underscore ( _ ) only**
- **Keywords Should not be used as Identifiers** 
- **White space is not allowed** in Identifiers 
- No Special characters are allowed **except an underscore ( _ )**
