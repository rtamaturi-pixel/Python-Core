## List comprehension:

1. **List comprehension** in the python is a concise and elegant way to create Lists.
2.  It allows us to create a new list by applying an expression to each item in a iterable (list, range, or string) - all in one single line.
   


```python
# wirte a programme to print all the even numbers in a range 1-20
```

```python
for i in range(1,21):
    if i % 2 == 0:
        print(i,end=' ')
```

```python
[i for i in range (1,21) if i%2==0]     #code using list comprehension.
```

```python
# Create a list of squares of numbers from 1 to 10.
```

```python
[ i ** 2 for i in range (1,11) ]
```

```python
# create a list of odd numbers in a range 1-30, using list comprehension.
```

```python
[i for i in range (1,31) if i % 2 != 0 ]
```

## Dictionary Comprahension:

1. Dictionary Comprahension is **compact and efficient ways** to create dictionaries from Iterables (List, Tuples, Range) in a **single line of code.**
2. It allows us to **Construct Dictionaries** quickly.
3. **Transform** keys and values.
4. **Filter** Items based on a condition.
5. Dictionary Comprehension, allows us to create Key-Value pair, in one neat line. Insted of writiing long for loops.

```python
squares = {}
for i in range(1,10):
    squares[i] = i **2
print(squares)
```

```python
{i:i**2 for i in range(1,10)}
```

```python
# create a dictionary of even numbers and thier squares in range 1 to 20
```

```python
{i:i**2 for i in range(1,21) if i%2==0}
```

```python
# create a dictionary of character frequencies in given string.
```

```python
S = 'programming'
```

```python
{i:S.count(i) for i in S}
```

```python
{i:S.count(i) for i in S}
```

```python
#without dictionary comprehension:

D = {}
for i in S:
    if i in D:
        D[i] += 1
    else:
        D[i] = 1
print(D)
    
    
```

```python
# using list comprehension:

[(i,S.count(i)) for i in S]
```

```python
set([(i,S.count(i)) for i in S])      #remove the repeated of words
```
