## Patterens: 

### Simple Right angle traiangle.

```python
n = 5
```

```python
for i in range (1,n+1):    #outer loop for rows.
    for j in range(i):     #inner loop for columns
        print("*", end = '')
    print()
```

```python
for i in range(1, n+1):   #other simple way.
    print("*" * i)
```

```python
for i in range (1, 11):
    for j in range(i):
        print("*", end='')
    print()
```

```python
# inverted right angle triangle.

n = 5   
```

```python
for i in range (n, 0, -1):    # here -1 is setp value, -1 * for every inner loop run.
    for j in range(i):     #inner loop for columns
        print("*", end = '')
    print()
```

### Number Patteren: 

```python
for i in range (1,n+1):    #outer loop for rows.
    for j in range(1, i+1):     #inner loop for columns
        print(j, end = '')
    print()
```

```python
for i in range(n, 0, -1):
    for j in range(1, i):
        print(j, end='')
    print()
```

### Pyramid Patteren:

```python
n = int(input("Enter the user number ="))

for i in range(1, n+1):
    print(" " * (n-i), end='')   # this print space x n-i times. end used so that it comes next line
    print("*"  * (2*i -1))  # you print * x mentioned number of times aligned in given space centered.
```

### inverted Pyramid patteren:

```python
n = int(input("Enter the user number ="))

for i in range(n, 0, -1):  #here n to 0, using -1 as step.
    print(" " * (n-i), end='')   # this print space x n-i times. end used so that it comes next line
    print("*"  * (2*i -1))  # you print * x mentioned number of times aligned in given space centered.
```
