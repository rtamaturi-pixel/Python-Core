### if-elif-else:

- **to check the multiple conditions in a code**
- if condition one is false, then check condition 2, if condition 2 is false, it goes to condition 3 and so on. till the default else statement at the end.
- If any condition is True, It exits there dont check others.

```python
num = int(input("Enter the number:"))

if num > 0:
    print("Given number is POSITIVE")
elif num < 0:
    print("Given number is NEGATIVE")
elif num ==0:
    print("Given number is ZERO")
else:
    print("Invalid input")
```

```python
a = int(input("Enter the value of a:"))
b = int(input("Enter the value of b:"))
choice = input("Enter the value of choice:")

if choice == "+":
    print("The result of Addition is :", a + b)
elif choice == "-":
    print("The result of Substraction is :", a - b)
elif choice == "*":
    print("The result of Multiplication is :", a * b)
elif choice == "/":
    print("The result of Division is :", a / b)
elif choice == "%":
    print("The result of Modulus is :", a % b)
elif choice == "//":
    print("The result of Floor Division is :", a // b)
elif choice == "**":
    print("The result of Power is :", a ** b)
else:
    print("Enter the valid number")
```

## Nested If:

- A nested if means an **if statement inside another if statement.**

```python
age = 20
if age > 10:              # first check
    if age > 18:          # second check inside the first
        print("You are an adult!")
```

Here:

- Python checks age > 10 → True

- Then it goes inside and checks age > 18 → True

- So it prints "You are an adult!"

- Make sure to check the indentation properly before exectue.

## Range: 

- In Python, range is a function that gives you a **sequence of numbers.**
- It is mostly used with loops (control flow) **to tell how many times the loop should run.**

```python
for i in range(5):  # we are telling programme to run 5-1 = 4 times
    print(i,end=' ')
```
