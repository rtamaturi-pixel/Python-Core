## Break :


- the Break statements is used to **exit a Loop immediately, when a certain condition is met.** It stops the current loop completely.

```python
for i in range (1,11):
    if i == 8:
       break
    print(i)
```

## Continue :
- The continue statement is used to **skip the current Iteration, of the loop and move to the next iteration** immediately.
- unlike break statement, it does not stop the loop Entirely, it just skip a trun.

#### using continue statement in for loop:

```python
countries = ["India", "Nepal", "butan", "Indonesia", "Japan"]

for country in countries:
    if country == "butan":
        continue             # skips the bhutan and continue the next iteration.
    print(country,end=' ')
```

#### using continue in while loop:

```python
i = 0
while i < 10:
    i = i+1
    if i==3:       #when i becomes 3 it skips.
        continue
    print(i,end=' ')
```

```python
lists = [10,20,30,40,50]

for i in lists:
    if i == 30:
        print("number found!")
        break
    print(i)
```

```python
number = [5,-3,8,-1,4]

for i in number:
    if i < 0:
        continue
    print(i)
```

```python
while True:
    n = int(input("Enter the number (0 to stop)"))
    if n == 0:
        print("stopping the input.....")
        break
    print("you entered: ", n)
```

```python
for i in range(1,11):
    if i % 3 ==0:
        continue
    print(i,end=' ')
```
