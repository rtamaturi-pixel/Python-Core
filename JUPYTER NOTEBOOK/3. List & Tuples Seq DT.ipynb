## LIST : [  ]
- A list is a collection that allows you to **store multiple items in a single variable**.
- It is written inside **square brackets [ ]**.
- it **hold different data types** (numbers, strings, booleans, even other lists).
- Lists are **ordered**
- Lists are **mutable** (you can change them)
- it **allow duplicates**.

## List Methods :

### append() :  one character (or) word can be inserted only at the end.

```python
fruits = ["apple","bannana","mango","orange"]
```

```python
fruits.append("kiwi")
```

```python
print(fruits)
```

```python
Enumbers = [2,4,6,8,10,12,14,16,18]
Enumbers.append(20)
print(Enumbers)
```

### extend() : multiple characters can be inserted only at end

```python
fruits = ["apple","bannana","mango","orange"]
print(fruits)
```

```python
fruits.extend(["kiwi","blueberry", "strawberry"])
```

```python
fruits
```

### Insert() : used to insert the element in a specific position.

```python
fruits = ["apple","bannana","mango","orange"]
```

```python
fruits.insert(1,"kiwi")  # 1 is the Index position to insert the element.
```

```python
print(fruits)
```

### remove() : incase if we have: removes the first matching value if we have duplicates.

```python
fruits.remove('apple')  #only one character or word can be removed.
```

```python
fruits
```

### pop() : remove the element with index number.

```python
fruits = ["apple", "bannana", "mango", "kiwi", "strawberry", "blueberry"]
```

```python
fruits
```

```python
fruits.pop(2) #index number 2 is 'mango', so it will pop and get removed!
```

```python
fruits
```

# TUPLE :  (   )

- A tuple is a collection of **ordered** items.

- Tuples are like lists, but **immutable** (you cannot change, add, or remove items once created).

- Written with **round brackets ( )**.

### Tuple Features :

1. Ordered → *items have a fixed order*.

2. Immutable → *cannot be modified after creation*.

3. Allows duplicates → *can store repeated values*.

4. Can store different data types → ***(1, "hello", 3.5)*** is valid.

### Tuple Methods () :
Since tuples are immutable, they have very few methods compared to lists.

### count() : returns how many times that value is appeared in Tuple

```python
a = (1,3,2,5,3,3,6,4,2,3)
```

```python
a.count(3)  #3 appeard 4 times in tuple a
```

```python
b = (2,4,6,"hello",2.3,45,"python","example",2,3,5,4,2,2,4,6)
```

```python
b.count("python")
```

```python
b.count(2)
```

### index() : Retruns the index number of the First value in occurance.

```python
a = ( 10,20,30,40,50,20,30,40,50,20,30,40,50)
```

```python
a.index(30) # 30 first occurance is in 2nd index.
```

### you can still use built-in functions with tuples, like:

**len(tuple)** → length

**min(tuple)** → smallest value

**max(tuple)** → largest value

**sum(tuple)** → sum of elements (if numbers)

**sorted(tuple)** → returns a sorted list
