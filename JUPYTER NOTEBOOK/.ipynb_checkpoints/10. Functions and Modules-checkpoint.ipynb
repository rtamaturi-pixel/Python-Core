### Function:

- function is a block of code, which **executes a pre-defined task**. **when ever it is called**.
- its key word is **def**
- set of instruction are called as programme. fuction is a type of programme, which execute a pre defined task.


#### Uses of Function:

- **Avoid repetition** – Write code once, reuse it.

- **Organize and optimize your program** – Make it easier to read and manage.

- **Debug easily** – If something goes wrong, it’s easier to find and fix.

- **Collaborate** – Different people can work on different functions.

#### Advantages:

- code reusablity
- code modularization
- code optimization

### Positional Arguments:
- The values are Passed **in an order**, and the **position matters**.
- the first values goes to the first paramenter, second value goes to the second paramenter and so on..

```python
def details(x,y):
    print("my name is:", x)
    print("my name is:", y)
    return " My name is "  + x +  " my age is "  + y    #space was given inside the " "
```

```python
details("Rupesh","26")
```

```python
# Design a simple arthematic caluclator using the Functions.
```

```python
def add(a,b):
    c = a + b
    return c

def sub(a,b):
    c = a - b
    return c

def mul(a,b):
    c = a * b
    return c

def div(a,b):
    c = a / b
    return c

def floor(a,b):
    c = a // b
    return c

def mod(a,b):
    c = a % b
    return c

def expo(a,b):
    c = a ** b
    return c
    
    
```

```python
a = int(input("Enter the a value:"))
b = int(input("Enter the b value:"))

choice = int(input("Enter choice of operation:"))

if choice == 1:
    print("The addition result is:",add(a,b))
elif choice == 2:
    print("The Substraction reuslt is:",sub(a,b))
elif choice == 3:
    print("The Multiplication result is:", mul(a,b))
elif choice == 4:
    print("The Division result is:", div(a,b))
elif choice == 5:
    print("The Floor division result is:", floor(a,b))
elif choice == 6:
    print("The Modulus result is:", mod(a,b))
elif choice == 7:
    print("The Exponential result is:", expo(a,b))
else:
    print("Enter the Valid choice of Operation for 1 to 7")
```

### Keyword Arguments:
- We Specify the **Parameter name, when passing the value.**
- Position does not matter here.
- Values are mapped based on Parameter name.

```python
def greetings(name,age,location,country,gender):
    print("my name is {}, My age is {}".format(name,age))
    print("my location is {}, My country is {}" .format(location,country))
    print("my Gender is {}" .format(gender))
    return name,age
```

```python
greetings(gender = 'male', country = 'India', age =26, location = 'hyd', name = 'Rupesh')
```

### Arbitary Arguments:
- When we dont know how many paramenter we pass to a function, then we take arbitary arguments.
- Arbitary arguments are denoted with * symbol.
- Arbitary arguments are collected as a Tuple of Items.

```python
def mydata(*args):
    print("my name is {}".format(args[0]))    # Indexing the Data
    print("my age is {}".format(args[1]))
    
```

```python
mydata('Rupesh',26,'India','Hyderabad')
```

```python
def data(*args):
    c = args[0] + args[1]
    d = args[2] - args[3]
    return c,d
    
```

```python
data(10,20,30,40,50,60,70,80)
```

### Arbitary Keyword Arguments:
- When we dont know, how many keyword arguments we pass to a function, then we take arbitary keyword arguments.
- Arbitary keyword arguments are **denoted by ** symbol**
- Arbitery keyword arguments are **collected as Key:Value** pair (dictionary)


```python
def data(**args):
    print("my name is:", args['name'])
    print("my age is:", args['age'])
    print("my location is:", args['location'])
```

```python
data(name = 'Rupesh', location = 'hyderabad', age = 26, country = 'India')
```
