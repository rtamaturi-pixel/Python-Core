## Object Oriented Programming: OOP's

- **Organize the code into Objects - is a instance of CLASS that combine data/attribute and behaviour/methods.**
- Object have its Own **Data/ATRIBUTES and Behaviours/METHODS** aspects.
- Python is a Object oriented programming Language.


#### example: Person (object)
- age, height, weight are properties / **attributes**
- walk, talk etc are behaviours / **methods**

#### Advantages:
- Code **reusablitiy** (We can reuse the code anytimes).
- More **security** (Encapsulation - we can limit its usablity to user).
- Code **Modularity** (we can divide code into small managable classes).
- Easy to maintain

#### Class: 
- a Class is a blue print or Template to create an object in python.
- It defines **Attributes (variable) and Methods(Functions)**, that describes the **state and Behaviour** of object.

- **Class** - Blue print or Template to create object.
- **Object** - Instance of Class (occured from class).
- **Attribute** - Data or features of Object.
- **Method** - Action of an Object.

#### Adding init() function to the Child class:

```python
class emp1(employee):
    def __init(self,name,age,designation,department,jobID,JD):
        self.JD = JD
```

- when we add init() function for the child class, the parent class init() will not be longer valid.
- if we want to have the parent class to along with child class init(), we need to use **super().** function.

```python
class emp1(employee):
    def __init(self,name,age,designation,department,jobID,JD):
        super().__init__(self,name,age,designation,department,jobID)     
```

#### super() function:
- it will inherit the properties of parent class to the child class where the child class contains __init()__ function

```python
class Employee:              #parent class
    def __init__(self,name,age,designation,department,JobId):
        self.name = name
        self.age = age 
        self.designation = designation
        self.department = department
        self.JobId = JobId
        
    def print_details(self):
        print('Employee name is: ',self.name)
        print('Employee age is: ',self.age)
        print('Employee Designation is: ',self.designation)
        print('Employee Department is: ',self.department)
        print('Employee Job Id is: ',self.JobId)
```

```python
class Emp1(Employee):            #child class
    def __init__(self,name,age,designation,department,JobId,JD,salary):
        super().__init__(name,age,designation,department,JobId)
        self.JD = JD
        self.salary = salary
```

#### Method Overriding

```python
class car:    #parent class
    def __init__(self,Model,Brand,Colour):
        self.Model = Model
        self.Brand = Brand
        self.Colour = Colour

    def print_car_details(self):
        print("The car Model is:", self.Model)
        print("The car Brand is:", self.Brand)
        print("The car Colour is:", self.Colour)
```

```python
class car1(car):    #child class
    def __init__(self,Model,Brand,Colour,YOM,Speed):
        super().__init__(Model,Brand,Colour)
        self.YOM = YOM
        self.Speed = Speed

    def print_car_details1(self):
        print("The Car YOM is:", self.YOM)
        print("The Car Speed is:", self.Speed)
        
```

```python
c1 = car1('220d',"BMW","Black",2025,"300km/hr")
```

#### Multi- Level Inheritance:

```python
class grandfather:
    def __init__(self, grand_bloodgroup, familyname):
        self.grand_bloodgroup = grand_bloodgroup
        self.familyname = familyname
```

```python
class father(grandfather):
    def __init__(self, grand_bloodgroup, familyname, father_height):
        self.father_height = father_height
        super().__init__(grand_bloodgroup, familyname)
```

```python
class son(father):
    def __init__(self, grand_bloodgroup, familyname, father_height, son_weight):
        self.son_weight = son_weight
        super().__init__(grand_bloodgroup, familyname, father_height)
```

#### class Polymorphism:

-Class polymorphism : We can have the **same Method name in Multiple classes**.

```python
class car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def move(self):                            #same method - move
        print("Car drives!!")

class boat:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def move(self):                            #same method - move
        print("boat sails!!")

class plane:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def move(self):                            #same method - move
        print("Plane flies!!")
        
```

```python
c1 = car("skoda", "ss0di")
b1 = boat("vikranth", 776)
p1 = plane("Boeing", 747)
```

```python
c1.move()
```

```python
b1.move()
```

```python
p1.move()
```

```python
for x in {c1,b1,p1}:   #here we can use (), {}, [] 
    x.move()
```

### Encapsulation:
- It is a process of **Binding(wrapping) the data(variable) and methods(functions)** that operates on that data into single unit - usually class.
- It helps us **Restrict Direct Access** to some of an objects data, which helps in **data protection and security.**

#### Access Modifiers:
1. **Public** -- Accessible to everyone.
2. **Protected** -- Accessible within the class and its subclass(_variable).
3. **Private** -- Accessible only withine the class(__variable).

```python
class encapsulation:
    def __init__(self):   # here we no need to write (self, xxx, yyy)
        self.public_variable = "I am a Public variable"
        self._protected_variable = "I am a Protected variable"    # _ before the vairable
        self.__private_variable = "I am Private variable"         # __ before the variable.

    def display(self):
        print(self.public_variable)
        print(self._protected_variable)
        print(self.__private_variable)
```

```python
class myclass:
    def __init__(self,name,age,networth):
        self.name = name             #public property
        self._age = age              #protected property
        self.__networth = networth   #private property 

    def display(self):
        print("My name is:", self.name)
        print("My age is :", self._age)
        print("My networth is:", self.__networth)
        
```

```python
data = myclass("Rupesh",26, 1000000)
```

```python
data.name
```

```python
data._age
```

```python
data.__networth    # it shows error, as networth is private.
```

```python
# we can only access the private object with the method():
```

```python
data.display()
```

#### Accessing Private variable using getter and setter:

```python
class bankaccount:
    def __init__(self,balance):
        self.__balance = balance   #private proterty

    # Getter Method:
    def get_balance(self):
        return self.__balance      # in retrun, Private property can not hide.

    # Setter Method:
    def set_balance(self, amount):
        if amount > 0:
            self.__balance = amount
        else:
            print("Invalid amount")
            
```

```python
acc = bankaccount(10000000)
```

```python
acc.balance    # we can not directly access, as it is private
```

```python
acc.get_balance()
```

```python
acc.set_balance(200000)
```

```python
acc.get_balance()
```

```python
acc.set_balance(-50090)
```

```python
class bankaccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder   #public property
        self.__balance = balance    #private property

    # public method to display balance (getter):\
    def get_balance(self):
        return self.__balance

    #public method to deposit the money (setter):
    def deposit(self,amount):
        if amount > 0:
            self.__balance += amount
            print(f"The deposited amount is: {amount}, and new balance is: {self.__balance}")
        else:
            print("Deposit amount should be positive")

        # Public method to with draw money:
    def withdraw(self,amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"withdraw: {amount}, Balance left is: {self.__balance}")
        else:
            print("Insufficient funds")
```

```python
# Creating amount object:
acc1 = bankaccount("Rupesh",1000000)
```

```python
#Accessing public variable:
acc1.account_holder
```

```python
#trying to access private property:
acc1.__balance
```

```python
# accessing private property using getter method:
acc1.get_balance()
```

```python
#deposit and withdraw operation:
acc1.deposit(20000)
```

```python
acc1.withdraw(30000)
```

```python
acc1.withdraw(10000000)
```

1. Encapsulation **prevents** Accidental Modification of Data.
2. It provides controlled acess through **getter** and **setter method**
3. Improve code Maintainability and **security.**

```python
class employee:
    def __init__(self,name,salary):
        self.name = name
        self._salary = salary   #protected

class manager(employee):
    def __init__(self,name,salary,department):
        super().__init__(name,salary)
        self.department = department

    #method inside a manager class:
    def show_manager_info(self):
        #accessing protected variables from parent class.
        print(self.name,self._salary, self.department)
```

```python
#create an object for subclass:
mngr = manager("Rupesh",80000,"HR")
```

```python
mngr.show_manager_info()
```

1. Protected variables in Python are **intended to be used with in its class and subclasses**.
but not **Accessed directlyt** from outside the class.

2. This is **only a convention**- It does not restrict the access fully

3. It **indicates to other programmers that the variable are for internal use only.**
