### Membership operator:

Membership operators are used to check if a value (element) exists in a sequence (like a list, string, tuple, set, or dictionary).

**Python has two membership operators:**

**in → returns True if the value is present.**

**not in → returns True if the value is NOT present.**

#### Membership in (in): 

- Returns True if Value is in the sequence.
- case sensitive

```python
fruits = ["apple","bannana","kiwi","watermelon"]
```

```python
"apple" in fruits
```

```python
"mango" in fruits
```

```python
set = (1,2,3,4,5,6)
```

```python
1 in set
```

```python
7 in set
```

#### Membership not in (not in) :

- Returns True if Value is NOT in the sequence.
- case sensitive

```python
fruits = ["apple","bannana","kiwi","watermelon"]
```

```python
"mango" not in fruits
```

```python
"kiwi" not in fruits
```

### Identity Operators:

Identity operators are used to check if **two variables refer (point) to the same object in memory.**

**Python has two identity operators:**

**is** → returns **True if both variables point to the same object.**

**is not** → returns **True if they point to different objects.**

#### Identity is (is):

```python
a = { 1, 2, 3, 4, 5, 6}
b = a
```

```python
a is b      # values are same, and memory is also same.
```

```python
b is a
```

```python
a = {1, 2, 3, 4, 5, 6}
b = {1, 2, 3, 4, 5, 6}
```

```python
a is b     # both are same as values, but different in Memory. so its false
```

```python
b is a     # diffenrent as the object memory is different.
```

#### identity is not (is not):

```python
a = {1,2,3,4,5,6}
b = a
```

```python
a is not b    # returns false as the both  are same.
```

```python
b is not a
```

```python
a = {1, 2, 3, 4, 5, 6}
b = {1, 2, 3, 4, 5, 6}
```

```python
b is not a   # returns True as the they are NOT same, object memory.
```

```python
a is not b
```

### Bitwise operator:

- Bit wise operator performs operation on **binary data**
1. Bitwise and **(&)**
2. Bitwise or
3. Bitwise xor
4. Bitwise not
5. bitwise left shift
6. Bitwise Right shift.

#### bitwise and (&):
 it returns True if both bits are true, false otherwise.

```python
a = 2
b = 3
```

```python
a & b 
```

```python
3 & 4
```

```python
4 & 9
```

#### Bitwise or (|) : 
It returns True, if even one bit is true, else false

```python
a = 2
b = 3
```

```python
a | b
```

```python
4 | 5
```

```python
5 | 9
```

#### Bitwise not: (~)

```python
a = 2
~a 
```

#### Bitwise xor: (^)
- It retruns True if One bit is True and One bit is Flase.
- 1 0 - True
- 0 1 - True
- Rest all false

```python
2 ^ 3
```

#### Bitwise Leftshift (<<):

**shift the bits positions to the left and takes the reusult.**

```python
a = 2
```

```python
a << 3  # here 3 is the no of bits you need to shift.
```

```python
a = 4
a << 2
```

#### Bitwise Rightshift (>>) :

**shift the bits positions to the Right and takes the reusult.**

```python
a = 8
```

```python
a >> 2
```

```python
a =6
```

```python
a >> 2
```
