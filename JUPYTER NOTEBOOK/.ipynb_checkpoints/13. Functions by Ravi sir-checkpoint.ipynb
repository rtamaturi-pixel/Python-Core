```python
#writing a function:

def welcome():
    print("Hello, World!")
    print("Today we are learning functions")
```

```python
# calling a function:

welcome()
```

```python
# defining a function with 'name' parameters:

def welcome(name):  #this function have one input parameter, so one o/p
    print(f"Hello, {name}")
    print("We are learning python functions")
```

```python
welcome("Rupesh")
```

```python
def welcome(name, age):
    print(f"Hello my name is {name}, my age is {age}.")
```

```python
welcome("Rupesh", 26)  
```

#### They are Positional:


```python
welcome(26, "Rupesh")
```

#### using named arguments to bypass positions:

```python
welcome(age=23, name="Rupesh")
```

```python
def product(name,price,discountpercent):
    print("Name:", name)
    print("Price:", price)
    print("discount% :", discountpercent)
    
```
