# SETS : {   }
## Set is a collection of Hetrogenious items, in unorder.

- It does not allow any **Duplicate or Repeated Values**.
- Stores unique items (**no duplicates allowed**).
- Is **unordered** (no fixed position or index).
- Uses **curly braces { }** or the set() function.


## Sets Methods: 

### union() :  Combine the total data, and merge the same data to single.

```python
S1 = {10,46,78,90,45,34,23,89,76,54}
S2 = {10,46,72,92,94,45,65,53,83}
```

```python
S1.union(S2)
```

```python
S3 = {'Python','4+8j'}
```

```python
S1.union(S2,S3)
```

### Intersection()   it is the common data in both sets.

```python
S1,S2
```

```python
S1.intersection(S2)
```

```python
print(help(set.intersection))
```

### difference() :  It deletes the Common data and give the rest of data.

```python
S1,S2
```

```python
S1.difference(S2)   # it is the remaining data in S1, after removing common data from both S1 and S2
```

```python
S2.difference(S1)   # it is the remaining data in S2, after removing common data from both S1 and S2
```

### issubset() - It is a True or false to check one set is the subset or derived from other set.


```python
print(help(set.issubset))
```

```python
S4 = {34,56,78,90,12,35,65,76,87}
S5 = {90,65,34}
```

```python
S5.issubset(S4)
```

## Dictionary {}

- It is a Collection of Data Items,each item is a ***KEY:VALUE*** pair.
- it is written is { } - curly brackets.
- Items can be accesed with thier Key names.
- Key Duplication is Not allowed.

```python
D = {'Brand' : 'BMW',
     'Model' : '220d',
     'YOM'   : 2025,
     'Colour': 'Phantom Black',
     'Speed' : '400km/hr' }
```

```python
D
```

```python
print(D)
```

```python
# to add one more Key and its value. " D[ key ] = value "
```

```python
D['fuel type'] = 'Petrol'
```

```python
D
```

```python
# Replace is with same key but different value, it overrides as the duplication is not allowed.
```

```python
D['Colour'] = 'Blue'
```

```python
D
```

```python
D['sunroof'] = 'yes'
```

```python
D
```
